```python
# prompt: QuesƟon 1
# A city planner is analyzing the number of vehicles owned by N households in an apartment
# complex (0 to 5 vehicles per household). The planner wants to understand the distribuƟon of
# vehicle ownership.
#  Generate a dataset represenƟng the number of vehicles owned per household.
#  Compute and plot the Probability Mass FuncƟon (PMF) and CumulaƟve DistribuƟon
# FuncƟon (CDF) to determine the probability of randomly selecƟng a household with a
# given number of vehicles.
# QuesƟon 2
# A factory produces a large number of gadgets, and each gadget has a 5% probability of being
# defecƟve. Given that a quality control team inspects N gadgets (N ≥ 100):
# i) What is the probability that exactly 10 gadgets are defecƟve?
# ii) What is the probability that at least 15 gadgets are defecƟve?
# iii) On average, how many defecƟve gadgets should the factory expect?

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Question 1: Vehicle Ownership Distribution

# Generate a dataset representing the number of vehicles owned per household
# Let's assume we have 100 households in the complex
n_households = 100
np.random.seed(42) #for reproducibility
vehicles_per_household = np.random.randint(0, 6, n_households) # 0 to 5 vehicles

# Compute the Probability Mass Function (PMF)
unique_vehicles, counts = np.unique(vehicles_per_household, return_counts=True)
pmf = counts / n_households

# Compute the Cumulative Distribution Function (CDF)
cdf = np.cumsum(pmf)


# Plot the PMF and CDF
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(unique_vehicles, pmf)
plt.title('Probability Mass Function (PMF)')
plt.xlabel('Number of Vehicles')
plt.ylabel('Probability')

plt.subplot(1, 2, 2)
plt.step(unique_vehicles, cdf)
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Number of Vehicles')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()

print("PMF:", pmf)
print("CDF:", cdf)


# Question 2: Defective Gadgets

# Number of gadgets inspected
n_gadgets = 200 # Example value (N >= 100)

# Probability of a gadget being defective
p_defective = 0.05


# i) Probability of exactly 10 defective gadgets
k_10 = 10
prob_exactly_10 = binom.pmf(k_10, n_gadgets, p_defective)
print(f"Probability of exactly 10 defective gadgets: {prob_exactly_10}")

# ii) Probability of at least 15 defective gadgets
k_15 = 15
prob_at_least_15 = 1 - binom.cdf(k_15 -1, n_gadgets, p_defective)
print(f"Probability of at least 15 defective gadgets: {prob_at_least_15}")


# iii) Expected number of defective gadgets
expected_defective = n_gadgets * p_defective
print(f"Expected number of defective gadgets: {expected_defective}")
```
