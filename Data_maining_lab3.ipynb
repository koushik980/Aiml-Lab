```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
data = pd.read_csv(url)
```

```python
print("Statistical Summary:")
print(data.describe())
```

```python
print("\nCategorical Summary:")
print(data.describe(include=['object']))
```

```python
plt.hist(data['Age'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['Pclass'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['Age'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
# 3. Data Visualizations using Seaborn
# Pair plot to visualize relationships between numerical features sns.pairplot(data[['Age', 'Fare', 'SibSp', 'Parch']].dropna()) plt.show()
```

```python
# Correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = data[['Age', 'Fare', 'SibSp', 'Parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count plot for 'Survived'
sns.countplot(x='Survived', data=data, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='Pclass', y='Age', data=data, palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.show()
```
