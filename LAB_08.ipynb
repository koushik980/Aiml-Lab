```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

# Data
study_hours = [2, 4, 6, 8, 1, 5, 7, 3, 9, 10]
exam_scores = [50, 60, 70, 85, 40, 65, 78, 55, 90, 95]
sleep_hours = [8, 7, 6, 5, 9, 6.5, 5.5, 7.5, 4, 3]

# 1. Scatter plots
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.scatter(study_hours, exam_scores)
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study Hours vs Exam Scores')

plt.subplot(1, 3, 2)
plt.scatter(sleep_hours, exam_scores)
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Scores')
plt.title('Sleep Hours vs Exam Scores')

plt.subplot(1, 3, 3)
plt.scatter(study_hours, sleep_hours)
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')
plt.title('Study Hours vs Sleep Hours')

plt.tight_layout()
plt.show()

# 2. Pearson Correlation Coefficients
corr1, _ = pearsonr(study_hours, exam_scores)
corr2, _ = pearsonr(sleep_hours, exam_scores)
corr3, _ = pearsonr(study_hours, sleep_hours)

print(f'Correlation between Study Hours and Exam Scores: {corr1:.2f}')
print(f'Correlation between Sleep Hours and Exam Scores: {corr2:.2f}')
print(f'Correlation between Study Hours and Sleep Hours: {corr3:.2f}')

# 3. Interpretation
if corr1 > 0:
    print("\nObservation: There is a positive correlation between Study Hours and Exam Scores.")
else:
    print("\nObservation: There is no significant positive correlation between Study Hours and Exam Scores.")
```
