```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data points
mean = 50
std_dev = 5
data = np.random.normal(mean, std_dev, 1000)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue')

# Overlay PDF
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r-', linewidth=2, label='PDF')

# Add labels and title
plt.title("Histogram and PDF of Normal Distribution")
plt.xlabel("Data Points")
plt.ylabel("Density")
plt.legend()
plt.grid(True)

# Show plot
plt.show()

import scipy.stats as stats
mean = 14
std_dev = 2.5
total_cases = 1000
z1 = (14 - mean) / std_dev  # For x = 14
z2 = (12 - mean) / std_dev  # For x = 12
z3 = (19 - mean) / std_dev  # For x = 19
z4 = (2.5- mean) / std_dev # For x = 2.5
prob1 = stats.norm.cdf(z1) - stats.norm.cdf(z4)
num_students1 = int(prob1 * total_cases)
prob2 = stats.norm.cdf(z2)
num_students2 = int(prob2 * total_cases)
prob3 = 1 - stats.norm.cdf(z3)
num_students3 = int(prob3 * total_cases)
print(f"I. Number of students scored between 14 and 2.5: {num_students1}")
print(f"II. Number of students scored below 12: {num_students2}")
print(f"III. Number of students scored above 19: {num_students3}")
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
plt.plot(x, stats.norm.pdf(x, mean, std_dev))
plt.fill_between(x, stats.norm.pdf(x, mean, std_dev), where=(x < 12), color='r', alpha=0.5)
plt.title('Probability of Scoring Below 12')
plt.xlabel('Score')
plt.ylabel('Probability Density')
plt.show()
```
